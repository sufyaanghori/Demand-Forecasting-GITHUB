```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# importing datasets
brazilian_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\Brazilian E-commerce Public Dataset by Olist\\final_brazilian_dataset.csv')
ecom_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\E-commerce Dataset\\final_E-commerce_dataset.csv')
wish_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\Sales of summer clothes in E-cmmerce Wish\\final_Wish_dataset.csv')
```

```python
# deleting the customer_id column
brazilian_df.drop('customer_id', axis=1, inplace=True)
```

```python
# deleting product_category_name column
brazilian_df.drop('product_category_name', axis=1, inplace=True)
```

```python
# purchase timestamp to date conversion
for date in brazilian_df['order_purchase_timestamp']:
    date = str(date)
    date_f = datetime.strptime(date, '%Y-%m-%d')
    date_f = date_f.strftime("%d-%m-%Y")

    brazilian_df['order_purchase_timestamp'] = brazilian_df['order_purchase_timestamp'].replace(date, date_f)
```

```python
# combining brazilian_df and ecom_df in one dataframe
brazilian_ecom_df = pd.concat([brazilian_df, ecom_df], axis=0)
brazilian_ecom_df['Order ID'] = brazilian_ecom_df['order_id'].fillna(brazilian_ecom_df['Order ID'])
brazilian_ecom_df.drop('order_id', axis=1, inplace=True)
```

```python
brazilian_ecom_df['Order Date'] = brazilian_ecom_df['order_purchase_timestamp'].fillna(brazilian_ecom_df['Order Date'])
brazilian_ecom_df['Category'] = brazilian_ecom_df['product_category_name_english'].fillna(brazilian_ecom_df['Category'])
brazilian_ecom_df['City'] = brazilian_ecom_df['customer_city'].fillna(brazilian_ecom_df['City'])

# drop the order_purchase_timestamp column
brazilian_ecom_df.drop('order_purchase_timestamp', axis=1, inplace=True)
brazilian_ecom_df.drop('product_category_name_english', axis=1, inplace=True)
brazilian_ecom_df.drop('customer_city', axis=1, inplace=True)
```

```python
# in quantity column, replace the NaN values with 1.0
brazilian_ecom_df['Quantity'] = brazilian_ecom_df['Quantity'].fillna(1.0)
```
