```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
order_list_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\E-commerce Dataset\\List of Orders.csv')
order_details_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\E-commerce Dataset\\Order Details.csv')
sales_target_df = pd.read_csv('C:\\Users\\sufya\\OneDrive\\Desktop\\Demand Forecasting\\E-commerce Dataset\\Sales Target.csv')
```

```python
# create a new dataframe df_ecommerce by merging the following:
# from order_list_df: 'Order ID', 'Order Date', 'CustomerName', 'City'
# from order_details_df: joining 'Order ID' with 'quantity', 'Category' and 'Sub-Category'

df_ecommerce = pd.merge(order_list_df[['Order ID', 'Order Date', 'CustomerName', 'City']],
                        order_details_df[['Order ID', 'Category', 'Sub-Category', 'Quantity']],
                        on='Order ID')
```
